# matplotlib simple plot 

```python
import matplotlib.pyplot as plt
import numpy as np
import math
a = np.arange(0, math.pi*2, 0.05)
b = np.sin(a)
plt.plot(a,b)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
a = np.arange(0, math.pi*1, 0.05)
b = np.tan(a)
plt.plot(a,b)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```
